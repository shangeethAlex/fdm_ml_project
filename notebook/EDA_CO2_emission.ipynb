```python

#conversion function
def l_100km_to_mpg(l_100km):
    if l_100km > 0:  # Ensure no division by zero or negative values
        return 282.481 / l_100km
    return np.nan

#use the function in the the data_prep
# Use .loc to avoid SettingWithCopyWarning
data_prep.loc[:, 'Fuel Consumption Comb (mpg)'] = data_prep.apply(
    lambda row: l_100km_to_mpg(row['Fuel Consumption Comb (L/100 km)']) 
    if pd.isna(row['Fuel Consumption Comb (mpg)']) and not pd.isna(row['Fuel Consumption Comb (L/100 km)']) 
    else row['Fuel Consumption Comb (mpg)'],
    axis=1
)
```

```python
data_prep[data_prep["Fuel Consumption Comb (mpg)"] < 0 ]
```

```python
data_prep.loc[
    data_prep['Fuel Consumption Comb (mpg)'] < 0, 
    'Fuel Consumption Comb (mpg)'
] = data_prep.apply(
    lambda row: l_100km_to_mpg(row['Fuel Consumption Comb (L/100 km)']) 
    if pd.isna(row['Fuel Consumption Comb (mpg)']) or row['Fuel Consumption Comb (mpg)'] < 0
    else row['Fuel Consumption Comb (mpg)'],
    axis=1
)
```
